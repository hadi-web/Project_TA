```python
# Stemming dengan NLTK (bahasa inggris)
from nltk.stem import *
from nltk.stem.porter import *

stemmer = PorterStemmer()
steming = [stemmer.stem(buku) for buku in output]
print(steming)
```

```python
import re
# numbering
numbering = [re.sub(r"\d+", "", buku) for buku in steming]
#punctuatuion
filtering_punctuation = [re.sub('[!"”#$%&’()*+,-./:;<=>?@[\]^_`{|}~]', ' ', buku) for buku in numbering]
data_filtering = pd.Series(filtering_punctuation)
data_filtering[:10]
```

```python
# tokenize
from nltk.tokenize import TweetTokenizer
tokens = TweetTokenizer()
data_tokenize = [tokens.tokenize(buku) for buku in data_filtering]
data_tokenize[:2]
```

# Pembuatan Model Naive Bayes

```python
# dengan data set terbaru
import pandas as pd
import nltk

buku_bersih = pd.read_csv('Data Bersih/data_bersih v3.csv')
buku_bersih.head()
```

```python
print(buku_bersih.groupby('Type Koleksi').size())
```

```python
print(buku_bersih['Type Koleksi'].unique())
```

```python
buku_bersih['everything'] = pd.DataFrame(buku_bersih['Judul'] + ' ' + buku_bersih['Penerbit'] + ' ' + buku_bersih['Tempat Terbit'] + ' ' + buku_bersih['Pengarang'])
print (buku_bersih['everything'].head())
```

## TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
tfidf_buku_bersih = tfidf_vectorizer.fit_transform(buku_bersih['everything'])
print(tfidf_buku_bersih)
```

```python
import csv
output_file = open('Data TF-IDF.csv', 'w')
datawriter = csv.writer(output_file)
datawriter.writerows(tfidf_buku_bersih)
output_file.close()
```

## Naive Bayes Classifier

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
```

```python
# create training and testing vars
#Importing train_test_split
from sklearn.model_selection import train_test_split
x_train  , x_test, y_train, y_test =train_test_split(tfidf_buku_bersih, buku_bersih['Type Koleksi'], test_size=0.1 ,random_state=15)

clf = MultinomialNB()
clf.fit(x_train,y_train)
Y_test_pred = clf.predict(x_test)
sklearn_score_train = clf.score(x_train,y_train)
print("Sklearn's score on training data :",sklearn_score_train)
sklearn_score_test1 = clf.score(x_test,y_test)
print("Sklearn's score on testing data :",sklearn_score_test1)
print("Classification report for testing data :-")
print(classification_report(y_test, Y_test_pred))
```
